# Matplotlib basics

<a href="http://matplotlib.org/">Matplotlib</a> es la librería de Python que emplearemos para ilustrar la realización de gráficos en 2 dimensiones con Python. Su programación es similar a Matlab y podremos generar gráficos en formatos PDF, JPG, SVG, PNG, BMP y GIF. 

En la Ciencia de datos tanto la realización de gráficos como la visualización tienen especial relevancia.

## Atributos y opciones en Matplotlib

Al emplear Matplotlib llamaremos al módulo <code>pyplot</code>. También podremos llamar a IPython para la realización de gráficos. 

```python
import matplotlib.pyplot as plt
import numpy as np

# Vector aleatorio con distribución gamma. 
L = np.random.gamma(0.5, 0.3, 50)

# Llamamos a IPython, si hemos instalado Anaconda no supondrá ningún problema
%matplotlib inline

# Deseamos graficar una línea con el vector de datos aleatorio
line, = plt.plot(L, "--")  

#Vamos a modificar la línea que queremos graficar
# Esa línea deseamos que tenga el color rojo
plt.setp(line, color="r")
#En este caso aplicamos una función sobre el objeto línea

# Cambiamos el grosor de la línea
line.set_linewidth(2.0)
#No hemos usado funciones, hemos modificado el atributo linewidth directamente sobre el objeto line

# Cambiamos algunas propiedades de la figura:
plt.title("Vector distribución gamma")
plt.ylabel("Valores gamma")
```

Otra forma de hacerlo:

```python
plt.plot(L, linestyle="--")
L2 = L + 0.1 
plt.plot(L2)
```

```python
plt.plot(L, linestyle="--")
L2 = L + 0.1 
plt.show()
```

## Modificar más elementos del gráfico ya generado 

Si estamos en el mismo Notebook podemos reutilizar un gráfico ya generado e ir añadiendo más elementos sobre él. En este caso **IPython genera en cada celda el gráfico** pero no es accesible para sucesivas partes del proceso.

```python
plt.title("título")
plt.show()
```

```python
plt.plot(L)
plt.show()
```

Cada celda conserva las características de un gráfico, si deseamos emplear un gráfico en las sucesivas celdas de nuestro notebook tendremos que hacer <code>set_matplotlib_close(False)</code> De este modo las características del gráfico se irán arrastrando a lo largo del notebook, esto será de especial utilidad cuando realicemos gráficos con alta complejidad. En este caso tenemos que trabajar con el módulo <code>set_matplotlib_close</code> que nos permite crear la figura y que no se pierda en otra celda del notebook.

Partimos creando la figura:

```python
from IPython.core.display import set_matplotlib_close

set_matplotlib_close(False)

plt.clf()
```

Simplemente añadimos un título:

```python
plt.title("título")
plt.show()
```

Sobre esa base añadimos las líneas:

```python
plt.plot(L)
plt.show()
```

Añadimos mayor complejidad. 

```python
plt.title("Nuevo título")
plt.plot([i+0.5 for i in L])
# Añadimos más líneas:
plt.show()
```

Podemos retornar al anterior comportamiento donde cada celda realizaría su propia figura.

```python
set_matplotlib_close(True)
plt.clf()
```

## Muchas figuras, el mismo gráfico

Podemos pintar múltiples líneas dentro del mismo gráfico.

```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline

# time serie 
t = np.arange(0., 5., 0.2)
print (t)

# Líneas rojas, cuadrados azules y triángulos verdes
lines = plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^')

print (len(lines))
```

Si empleamos los argumentos de <code>plt.plot</code> se lo aplicaríamos a todo el gráfico:

```python
plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^', color="g")
```

## Múltiples gráficos en una figura. Los subplot

Si empleamos  <code>subplot</code> podemos incluir varios gráficos en la misma figura.

```python
plt.figure(1)  # Crea la figura, se puede omitir.
plt.subplot(224) # 2 filas, 2 columnas, seleccionamos la última.

# Datos aleatorios:
data = np.random.random(100)
x = range(len(data))
plt.plot(x, data);

# Seleccionamos la xx1 la primera parte de la cuadrícula generada:
plt.subplot(221)
plt.scatter(x, data);
```

Otro ejemplo de gráfico 

```python
x1 = np.linspace(0.0, 5.0)
x2 = np.linspace(0.0, 2.0)

y1 = np.cos(2 * np.pi * x1) * np.exp(-x1)
y2 = np.cos(2 * np.pi * x2)

plt.subplot(2, 1, 1)
plt.plot(x1, y1, 'o-')
plt.title('A tale of 2 subplots')
plt.ylabel('Damped oscillation')

plt.subplot(2, 1, 2)
plt.plot(x2, y2, '.-')
plt.xlabel('time (s)')
plt.ylabel('Undamped')

plt.show()
```

Como hemos empleado <code>set_matplotlib_close(False)</code> estamos en disposición de modificar las propiedades de la figura en distintas celdas del notebook. En caso contrario lo haremos sobre la misma celda:

```python
set_matplotlib_close(True)

fig2 = plt.figure()
ax2 = fig2.add_subplot(1,1,1) # una fila, una columna
print (type(ax2))
ax2.hist([(i+1)**2 for i in L], bins=100)
ax2.set_title("Modificación sobre la Gamma")
```

Depende del momento en el que ejecutemos la anterior celda obtendremos un resultado u otro.

Una figura está compuesta de subplot, ejes,... y cada elemento tiene sus propiedades correspondientes. Mediante set_XX podemos modificar las propiedades del objeto matplotlib.

```python
fig3 = plt.figure()

points = fig3.add_subplot(1,2,1) # una fila dos columnas, empezamos en la primera
points.plot(L)
points.set_title("Puntos")

# Two ways of setting properties:
points.lines[0].set_marker('o')
plt.setp(points.lines[0], color="r")

histogram = fig3.add_subplot(1,2,2) # una fila dos columnas, seleccionamos la segunda
histogram.hist(L)
histogram.set_title("Histograma")
```

En un sólo gráfico podemos incluir 2 ejes y cada uno de ellos será un tipo de gráfico distinto. Veamos un ejemplo más concreto que además nos sirve para ilustrar otro modo de obtener conjuntos de datos de la web:

```python
import io 
import requests
#Lectura de un data set con número de siniestros de una cartera de automóviles
url = 'http://www.businessandeconomics.mq.edu.au/our_departments/Applied_Finance_and_Actuarial_Studies/acst_docs/glms_for_insurance_data/data/claimslong.csv'
s = requests.get(url).content
df = pd.read_csv(io.StringIO(s.decode('utf-8')))

df.head()
```

Mediante IO y requests nos descargamos un csv de la web que nos permite crear un data frame. En el eje principal vamos a graficar la exposición y en el otro eje vamos a graficar la frecuencia siniestral por año:

```python
frecuencia =  pd.DataFrame((df['claim']).groupby(df['period']).mean())
exposicion = pd.DataFrame((df['claim']).groupby(df['period']).count())
```

```python
fig = plt.figure()
#En el eje principal ponemos la exposición como barras
ax = exposicion['claim'].plot(kind='bar',grid=True)
#Añadimos el eje secundario
ax2 = ax.twinx()
#En este eje secundario ponemos la frecuencia como un gráfico de líneas
ax2.plot(frecuencia['claim'].values, linestyle='-', linewidth=2.0,color='red')
plt.show();
```
